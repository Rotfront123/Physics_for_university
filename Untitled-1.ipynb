```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_trajectory(model_type, v0, angle, k=0.1, dt=0.01, max_time=10):
    g = 9.81
    angle_rad = np.radians(angle)
    
    x, y = 0, 0
    vx = v0 * np.cos(angle_rad)
    vy = v0 * np.sin(angle_rad)
    
    x_points, y_points = [x], [y]
    
    t = 0
    while y >= 0 and t < max_time:
        v = np.sqrt(vx**2 + vy**2)
        
        if model_type == 'linear':
            F_res = k * v
        else: 
            F_res = k * v**2
     
        ax = -F_res * vx / v 
        ay = -g - F_res * vy / v 
        
        vx += ax * dt
        vy += ay * dt
        x += vx * dt
        y += vy * dt
        
        x_points.append(x)
        y_points.append(y)
        t += dt
    
    return x_points, y_points, x_points[-1]

def plot_trajectories(cases):
    plt.figure(figsize=(12, 8))
    for i, (model_type, v0, angle, k, label) in enumerate(cases):
        x, y, fall_point = simulate_trajectory(model_type, v0, angle, k)
        plt.plot(x, y, label=f'{label}', linewidth=2)
        plt.plot(fall_point, 0, 'o', markersize=8, label=f'ĞŸĞ°Ğ´ĞµĞ½Ğ¸Ğµ: {fall_point:.1f} Ğ¼')
    
    plt.xlabel('Ğ Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ, Ğ¼', fontsize=12)
    plt.ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼', fontsize=12)
    plt.title('Ğ¢Ñ€Ğ°ĞµĞºÑ‚Ğ¾Ñ€Ğ¸Ğ¸ Ğ´Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ñ ĞºĞ°Ğ¼Ğ½Ñ Ñ ÑƒÑ‡ĞµÑ‚Ğ¾Ğ¼ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ Ğ²Ğ¾Ğ·Ğ´ÑƒÑ…Ğ°', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axis('equal') 
    plt.tight_layout()
    plt.show()

def get_user_input():
    print("=== ĞœĞ¾Ğ´ĞµĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ´Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ñ ĞºĞ°Ğ¼Ğ½Ñ ===")
    print("\nĞ’Ñ‹Ğ±ĞµÑ€Ğ¸Ñ‚Ğµ Ñ‚Ğ¸Ğ¿ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ Ğ²Ğ¾Ğ·Ğ´ÑƒÑ…Ğ°:")
    print("1 - Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ (ğ¹ âˆ¼ ğ‘£)")
    print("2 - ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ (ğ¹ âˆ¼ ğ‘£Â²)")
    print("3 - Ğ¡Ñ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ğµ Ğ¾Ğ±Ğ¾Ğ¸Ñ… Ñ‚Ğ¸Ğ¿Ğ¾Ğ²")
    
    choice = input("Ğ’Ğ°Ñˆ Ğ²Ñ‹Ğ±Ğ¾Ñ€ (1-3): ").strip()
    
    if choice == '1':
        model_type = 'linear'
        k_default = 0.1
    elif choice == '2':
        model_type = 'quadratic'
        k_default = 0.01
    elif choice == '3':
        return 'compare'
    
    try:
        v0 = float(input(f"ĞĞ°Ñ‡Ğ°Ğ»ÑŒĞ½Ğ°Ñ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ (Ğ¼/Ñ) [Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ 20]: ") or 20)
        angle = float(input(f"Ğ£Ğ³Ğ¾Ğ» Ğ±Ñ€Ğ¾ÑĞºĞ° (Ğ³Ñ€Ğ°Ğ´ÑƒÑÑ‹) [Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ 45]: ") or 45)
        k = float(input(f"ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ {k_default}]: ") or k_default)
    except ValueError:
        print("ĞÑˆĞ¸Ğ±ĞºĞ° Ğ²Ğ²Ğ¾Ğ´Ğ°, Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµĞ¼ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ñ Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ")
        v0, angle, k = 20, 45, k_default
    
    return model_type, v0, angle, k

def single_simulation():
    params = get_user_input()
    
    if params == 'compare':
        cases = [
            ('linear', 20, 45, 0.1, 'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k=0.1)'),
            ('quadratic', 20, 45, 0.01, 'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k=0.01)'),
            ('linear', 20, 45, 0, 'Ğ‘ĞµĞ· ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ')
        ]
    else:
        model_type, v0, angle, k = params
        cases = [
            (model_type, v0, angle, k, f'ĞŸĞ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ñ‚ĞµĞ»ÑŒÑĞºĞ¸Ğµ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ñ‹'),
            (model_type, v0, angle, 0, f'Ğ‘ĞµĞ· ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ (Ğ´Ğ»Ñ ÑÑ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ñ)')
        ]
    
    plot_trajectories(cases)
    print("\n=== Ğ ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ñ‹ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ñ ===")
    for model_type, v0, angle, k, label in cases:
        x, y, fall_point = simulate_trajectory(model_type, v0, angle, k)
        print(f"{label}:")
        print(f"  - Ğ”Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ Ğ¿Ğ¾Ğ»ĞµÑ‚Ğ°: {fall_point:.2f} Ğ¼")
        print(f"  - ĞœĞ°ĞºÑĞ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ğ°Ñ Ğ²Ñ‹ÑĞ¾Ñ‚Ğ°: {max(y):.2f} Ğ¼")
        print(f"  - Ğ’Ñ€ĞµĞ¼Ñ Ğ¿Ğ¾Ğ»ĞµÑ‚Ğ°: {len(x)*0.01:.2f} Ñ")
        print()

def parameter_study():
    print("\n=== Ğ˜ÑÑĞ»ĞµĞ´Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ²Ğ»Ğ¸ÑĞ½Ğ¸Ñ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ¾Ğ² ===")
    print("Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑƒĞ³Ğ»Ğ° Ğ±Ñ€Ğ¾ÑĞºĞ° (v0=20 Ğ¼/Ñ, k=0.01):")
    angles = [30, 45, 60, 75]
    for angle in angles:
        x, y, fall_point = simulate_trajectory('quadratic', 20, angle, 0.01)
        print(f"  Ğ£Ğ³Ğ¾Ğ» {angle}Â°: Ğ´Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ = {fall_point:.1f} Ğ¼, Ğ²Ñ‹ÑĞ¾Ñ‚Ğ° = {max(y):.1f} Ğ¼")
    
    print("\nĞ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ Ğ½Ğ°Ñ‡Ğ°Ğ»ÑŒĞ½Ğ¾Ğ¹ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚Ğ¸ (ÑƒĞ³Ğ¾Ğ»=45Â°, k=0.01):")
    velocities = [10, 20, 30, 40]
    for v0 in velocities:
        x, y, fall_point = simulate_trajectory('quadratic', v0, 45, 0.01)
        print(f"  Ğ¡ĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ {v0} Ğ¼/Ñ: Ğ´Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ = {fall_point:.1f} Ğ¼, Ğ²Ñ‹ÑĞ¾Ñ‚Ğ° = {max(y):.1f} Ğ¼")

def main():
    while True:
        print("\n" + "="*50)
        print("1 - ĞĞ´Ğ¸Ğ½Ğ¾Ñ‡Ğ½Ğ¾Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ")
        print("2 - Ğ˜ÑÑĞ»ĞµĞ´Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ¾Ğ²")
        print("3 - Ğ’Ñ‹Ñ…Ğ¾Ğ´")
        
        choice = input("Ğ’Ñ‹Ğ±ĞµÑ€Ğ¸Ñ‚Ğµ Ğ´ĞµĞ¹ÑÑ‚Ğ²Ğ¸Ğµ (1-3): ").strip()
        
        if choice == '1':
            single_simulation()
        elif choice == '2':
            parameter_study()
        elif choice == '3':
            print("Ğ’Ñ‹Ñ…Ğ¾Ğ´ Ğ¸Ğ· Ğ¿Ñ€Ğ¾Ğ³Ñ€Ğ°Ğ¼Ğ¼Ñ‹...")
            break
        else:
            print("ĞĞµĞ²ĞµÑ€Ğ½Ñ‹Ğ¹ Ğ²Ñ‹Ğ±Ğ¾Ñ€, Ğ¿Ğ¾Ğ¿Ñ€Ğ¾Ğ±ÑƒĞ¹Ñ‚Ğµ ÑĞ½Ğ¾Ğ²Ğ°")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_trajectory(model_type, v0, angle, k=0.0, dt=0.01, max_time=20):
    g = 9.81
    angle_rad = np.radians(angle)
    x, y = 0, 0
    vx = v0 * np.cos(angle_rad)
    vy = v0 * np.sin(angle_rad)
    
    x_points, y_points = [x], [y]
    t_points = [0]
    t = 0
    while y >= 0 and t < max_time:
        v = np.sqrt(vx**2 + vy**2)
        if model_type == 'linear':
            F_res = k * v
        else:
            F_res = k * v**2
        ax = -F_res * vx / v if v > 0 else 0
        ay = -g - F_res * vy / v if v > 0 else -g
        vx += ax * dt
        vy += ay * dt
        x += vx * dt
        y += vy * dt
        t += dt
        
        x_points.append(x)
        y_points.append(y)
        t_points.append(t)
    
    return x_points, y_points, t_points, x_points[-1]

def theoretical_no_resistance(v0, angle, g=9.81):
    """Ğ¢ĞµĞ¾Ñ€ĞµÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¸Ğ¹ Ñ€Ğ°ÑÑ‡ĞµÑ‚ Ğ±ĞµĞ· ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ Ğ²Ğ¾Ğ·Ğ´ÑƒÑ…Ğ°"""
    angle_rad = np.radians(angle)
    
    range_val = (v0**2 * np.sin(2*angle_rad)) / g
    
    max_height = (v0**2 * (np.sin(angle_rad))**2) / (2*g)
    
    time_of_flight = (2 * v0 * np.sin(angle_rad)) / g
    
    return range_val, max_height, time_of_flight

def plot_angle_comparison():
    """Ğ ĞµĞ¶Ğ¸Ğ¼ 1: Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑƒĞ³Ğ»Ğ° Ğ½Ğ° Ğ±Ñ€Ğ¾ÑĞ¾Ğº"""
    print("\n=== Ğ ĞµĞ¶Ğ¸Ğ¼: Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑƒĞ³Ğ»Ğ° Ğ½Ğ° Ğ±Ñ€Ğ¾ÑĞ¾Ğº ===")
    
    try:
        v0 = float(input("ĞĞ°Ñ‡Ğ°Ğ»ÑŒĞ½Ğ°Ñ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ (Ğ¼/Ñ) [20]: ") or 20)
        k_linear = float(input("ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ Ğ»Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [0.1]: ") or 0.1)
        k_quadratic = float(input("ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [0.01]: ") or 0.01)
    except ValueError:
        print("ĞÑˆĞ¸Ğ±ĞºĞ° Ğ²Ğ²Ğ¾Ğ´Ğ°, Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµĞ¼ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ñ Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ")
        v0, k_linear, k_quadratic = 20, 0.1, 0.01
    
    angles = [30, 45, 60, 75]
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    for angle in angles:
        x, y, t, fall_point = simulate_trajectory('linear', v0, angle, k_linear)
        ax1.plot(x, y, label=f'{angle}Â°', linewidth=2)
        ax2.plot(t[:-1], [y[i] for i in range(len(y)-1)], label=f'{angle}Â°', linewidth=2)
    
    for angle in angles:
        x, y, t, fall_point = simulate_trajectory('quadratic', v0, angle, k_quadratic)
        ax3.plot(x, y, label=f'{angle}Â°', linewidth=2)
        ax4.plot(t[:-1], [y[i] for i in range(len(y)-1)], label=f'{angle}Â°', linewidth=2)
    
    ax1.set_title(f'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_linear})')
    ax1.set_xlabel('Ğ Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ, Ğ¼')
    ax1.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    ax2.set_title(f'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_linear})')
    ax2.set_xlabel('Ğ’Ñ€ĞµĞ¼Ñ, Ñ')
    ax2.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    ax3.set_title(f'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_quadratic})')
    ax3.set_xlabel('Ğ Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ, Ğ¼')
    ax3.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax3.grid(True, alpha=0.3)
    ax3.legend()
    
    ax4.set_title(f'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_quadratic})')
    ax4.set_xlabel('Ğ’Ñ€ĞµĞ¼Ñ, Ñ')
    ax4.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax4.grid(True, alpha=0.3)
    ax4.legend()
    
    plt.tight_layout()
    plt.show()
   
    print("\nĞ ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ñ‹ Ğ´Ğ»Ñ Ğ»Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ:")
    for angle in angles:
        x, y, t, fall_point = simulate_trajectory('linear', v0, angle, k_linear)
        range_theor, height_theor, time_theor = theoretical_no_resistance(v0, angle)
        print(f"Ğ£Ğ³Ğ¾Ğ» {angle}Â°: Ğ´Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ = {fall_point:.1f} Ğ¼ (Ñ‚ĞµĞ¾Ñ€. {range_theor:.1f} Ğ¼), "
              f"Ğ²Ñ‹ÑĞ¾Ñ‚Ğ° = {max(y):.1f} Ğ¼ (Ñ‚ĞµĞ¾Ñ€. {height_theor:.1f} Ğ¼)")

def plot_velocity_comparison():
    """Ğ ĞµĞ¶Ğ¸Ğ¼ 1: Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚Ğ¸ Ğ½Ğ° Ğ±Ñ€Ğ¾ÑĞ¾Ğº"""
    print("\n=== Ğ ĞµĞ¶Ğ¸Ğ¼: Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚Ğ¸ Ğ½Ğ° Ğ±Ñ€Ğ¾ÑĞ¾Ğº ===")
    
    try:
        angle = float(input("Ğ£Ğ³Ğ¾Ğ» Ğ±Ñ€Ğ¾ÑĞºĞ° (Ğ³Ñ€Ğ°Ğ´ÑƒÑÑ‹) [45]: ") or 45)
        k_linear = float(input("ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ Ğ»Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [0.1]: ") or 0.1)
        k_quadratic = float(input("ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [0.01]: ") or 0.01)
    except ValueError:
        print("ĞÑˆĞ¸Ğ±ĞºĞ° Ğ²Ğ²Ğ¾Ğ´Ğ°, Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµĞ¼ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ñ Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ")
        angle, k_linear, k_quadratic = 45, 0.1, 0.01
    
    velocities = [10, 20, 30, 40]
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    for v0 in velocities:
        x, y, t, fall_point = simulate_trajectory('linear', v0, angle, k_linear)
        ax1.plot(x, y, label=f'{v0} Ğ¼/Ñ', linewidth=2)
        ax2.plot(t[:-1], [y[i] for i in range(len(y)-1)], label=f'{v0} Ğ¼/Ñ', linewidth=2)
    
    for v0 in velocities:
        x, y, t, fall_point = simulate_trajectory('quadratic', v0, angle, k_quadratic)
        ax3.plot(x, y, label=f'{v0} Ğ¼/Ñ', linewidth=2)
        ax4.plot(t[:-1], [y[i] for i in range(len(y)-1)], label=f'{v0} Ğ¼/Ñ', linewidth=2)
    
    ax1.set_title(f'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_linear})')
    ax1.set_xlabel('Ğ Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ, Ğ¼')
    ax1.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    ax2.set_title(f'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_linear})')
    ax2.set_xlabel('Ğ’Ñ€ĞµĞ¼Ñ, Ñ')
    ax2.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    ax3.set_title(f'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_quadratic})')
    ax3.set_xlabel('Ğ Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ, Ğ¼')
    ax3.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax3.grid(True, alpha=0.3)
    ax3.legend()
    
    ax4.set_title(f'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_quadratic})')
    ax4.set_xlabel('Ğ’Ñ€ĞµĞ¼Ñ, Ñ')
    ax4.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax4.grid(True, alpha=0.3)
    ax4.legend()
    
    plt.tight_layout()
    plt.show()
    
    print("\nĞ ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ñ‹ Ğ´Ğ»Ñ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ:")
    for v0 in velocities:
        x, y, t, fall_point = simulate_trajectory('quadratic', v0, angle, k_quadratic)
        range_theor, height_theor, time_theor = theoretical_no_resistance(v0, angle)
        print(f"Ğ¡ĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ {v0} Ğ¼/Ñ: Ğ´Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ = {fall_point:.1f} Ğ¼ (Ñ‚ĞµĞ¾Ñ€. {range_theor:.1f} Ğ¼), "
              f"Ğ²Ñ‹ÑĞ¾Ñ‚Ğ° = {max(y):.1f} Ğ¼ (Ñ‚ĞµĞ¾Ñ€. {height_theor:.1f} Ğ¼)")

def personal_simulation():
    """Ğ ĞµĞ¶Ğ¸Ğ¼ 2: ĞŸĞµÑ€ÑĞ¾Ğ½Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ"""
    print("\n=== Ğ ĞµĞ¶Ğ¸Ğ¼: ĞŸĞµÑ€ÑĞ¾Ğ½Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ ===")
    
    try:
        v0 = float(input("ĞĞ°Ñ‡Ğ°Ğ»ÑŒĞ½Ğ°Ñ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ (Ğ¼/Ñ): "))
        angle = float(input("Ğ£Ğ³Ğ¾Ğ» Ğ±Ñ€Ğ¾ÑĞºĞ° (Ğ³Ñ€Ğ°Ğ´ÑƒÑÑ‹): "))
        k_linear = float(input("ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ Ğ»Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [0.1]: ") or 0.1)
        k_quadratic = float(input("ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğ³Ğ¾ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ [0.01]: ") or 0.01)
    except ValueError:
        print("ĞÑˆĞ¸Ğ±ĞºĞ° Ğ²Ğ²Ğ¾Ğ´Ğ°!")
        return
    
    range_theor, height_theor, time_theor = theoretical_no_resistance(v0, angle)
    
    x_linear, y_linear, t_linear, fall_linear = simulate_trajectory('linear', v0, angle, k_linear)
    x_quadratic, y_quadratic, t_quadratic, fall_quadratic = simulate_trajectory('quadratic', v0, angle, k_quadratic)
    x_no_res, y_no_res, t_no_res, fall_no_res = simulate_trajectory('linear', v0, angle, 0)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    ax1.plot(x_linear, y_linear, 'b-', label=f'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ (k={k_linear})', linewidth=2)
    ax1.plot(x_quadratic, y_quadratic, 'r-', label=f'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ (k={k_quadratic})', linewidth=2)
    ax1.plot(x_no_res, y_no_res, 'g--', label='Ğ‘ĞµĞ· ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ', linewidth=2)
    ax1.plot(fall_linear, 0, 'bo', markersize=8)
    ax1.plot(fall_quadratic, 0, 'ro', markersize=8)
    ax1.plot(fall_no_res, 0, 'go', markersize=8)
    
    ax1.set_xlabel('Ğ Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ, Ğ¼')
    ax1.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax1.set_title('Ğ¡Ñ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ğµ Ñ‚Ñ€Ğ°ĞµĞºÑ‚Ğ¾Ñ€Ğ¸Ğ¹')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    ax1.axis('equal')
    
    ax2.plot(t_linear[:-1], [y_linear[i] for i in range(len(y_linear)-1)], 'b-', 
             label=f'Ğ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ (k={k_linear})', linewidth=2)
    ax2.plot(t_quadratic[:-1], [y_quadratic[i] for i in range(len(y_quadratic)-1)], 'r-', 
             label=f'ĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ (k={k_quadratic})', linewidth=2)
    ax2.plot(t_no_res[:-1], [y_no_res[i] for i in range(len(y_no_res)-1)], 'g--', 
             label='Ğ‘ĞµĞ· ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ', linewidth=2)
    
    ax2.set_xlabel('Ğ’Ñ€ĞµĞ¼Ñ, Ñ')
    ax2.set_ylabel('Ğ’Ñ‹ÑĞ¾Ñ‚Ğ°, Ğ¼')
    ax2.set_title('Ğ—Ğ°Ğ²Ğ¸ÑĞ¸Ğ¼Ğ¾ÑÑ‚ÑŒ Ğ²Ñ‹ÑĞ¾Ñ‚Ñ‹ Ğ¾Ñ‚ Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ¸')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

    print("\n=== Ğ Ğ•Ğ—Ğ£Ğ›Ğ¬Ğ¢ĞĞ¢Ğ« ĞœĞĞ”Ğ•Ğ›Ğ˜Ğ ĞĞ’ĞĞĞ˜Ğ¯ ===")
    print(f"ĞŸĞ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ñ‹: v0 = {v0} Ğ¼/Ñ, ÑƒĞ³Ğ¾Ğ» = {angle}Â°")
    print(f"Ğ¢ĞµĞ¾Ñ€ĞµÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¸ (Ğ±ĞµĞ· ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ñ):")
    print(f"  Ğ”Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ: {range_theor:.2f} Ğ¼")
    print(f"  ĞœĞ°ĞºÑ. Ğ²Ñ‹ÑĞ¾Ñ‚Ğ°: {height_theor:.2f} Ğ¼")
    print(f"  Ğ’Ñ€ĞµĞ¼Ñ Ğ¿Ğ¾Ğ»ĞµÑ‚Ğ°: {time_theor:.2f} Ñ")
    
    print(f"\nĞ›Ğ¸Ğ½ĞµĞ¹Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_linear}):")
    print(f"  Ğ”Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ: {fall_linear:.2f} Ğ¼ ({fall_linear/range_theor*100:.1f}% Ğ¾Ñ‚ Ñ‚ĞµĞ¾Ñ€ĞµÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¾Ğ¹)")
    print(f"  ĞœĞ°ĞºÑ. Ğ²Ñ‹ÑĞ¾Ñ‚Ğ°: {max(y_linear):.2f} Ğ¼ ({max(y_linear)/height_theor*100:.1f}% Ğ¾Ñ‚ Ñ‚ĞµĞ¾Ñ€ĞµÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¾Ğ¹)")
    print(f"  Ğ’Ñ€ĞµĞ¼Ñ Ğ¿Ğ¾Ğ»ĞµÑ‚Ğ°: {t_linear[-1]:.2f} Ñ")
    
    print(f"\nĞšĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¸Ñ‡Ğ½Ğ¾Ğµ ÑĞ¾Ğ¿Ñ€Ğ¾Ñ‚Ğ¸Ğ²Ğ»ĞµĞ½Ğ¸Ğµ (k={k_quadratic}):")
    print(f"  Ğ”Ğ°Ğ»ÑŒĞ½Ğ¾ÑÑ‚ÑŒ: {fall_quadratic:.2f} Ğ¼ ({fall_quadratic/range_theor*100:.1f}% Ğ¾Ñ‚ Ñ‚ĞµĞ¾Ñ€ĞµÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¾Ğ¹)")
    print(f"  ĞœĞ°ĞºÑ. Ğ²Ñ‹ÑĞ¾Ñ‚Ğ°: {max(y_quadratic):.2f} Ğ¼ ({max(y_quadratic)/height_theor*100:.1f}% Ğ¾Ñ‚ Ñ‚ĞµĞ¾Ñ€ĞµÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¾Ğ¹)")
    print(f"  Ğ’Ñ€ĞµĞ¼Ñ Ğ¿Ğ¾Ğ»ĞµÑ‚Ğ°: {t_quadratic[-1]:.2f} Ñ")

def main():
    """Ğ“Ğ»Ğ°Ğ²Ğ½Ğ¾Ğµ Ğ¼ĞµĞ½Ñ Ğ¿Ñ€Ğ¾Ğ³Ñ€Ğ°Ğ¼Ğ¼Ñ‹"""
    while True:
        print("\n" + "="*60)
        print("           ĞœĞĞ”Ğ•Ğ›Ğ˜Ğ ĞĞ’ĞĞĞ˜Ğ• Ğ”Ğ’Ğ˜Ğ–Ğ•ĞĞ˜Ğ¯ ĞšĞĞœĞĞ¯")
        print("="*60)
        print("\nĞ ĞµĞ¶Ğ¸Ğ¼ 1: Ğ˜ÑÑĞ»ĞµĞ´Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ¾Ğ²")
        print("  1.1 - Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑƒĞ³Ğ»Ğ° Ğ½Ğ° Ğ±Ñ€Ğ¾ÑĞ¾Ğº")
        print("  1.2 - Ğ’Ğ»Ğ¸ÑĞ½Ğ¸Ğµ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚Ğ¸ Ğ½Ğ° Ğ±Ñ€Ğ¾ÑĞ¾Ğº")
        print("\nĞ ĞµĞ¶Ğ¸Ğ¼ 2: ĞŸĞµÑ€ÑĞ¾Ğ½Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ")
        print("  2 - Ğ’Ğ²Ğ¾Ğ´ Ğ¿Ğ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ñ‚ĞµĞ»ÑŒÑĞºĞ¸Ñ… Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ¾Ğ²")
        print("\n  0 - Ğ’Ñ‹Ñ…Ğ¾Ğ´")
        
        choice = input("\nĞ’Ñ‹Ğ±ĞµÑ€Ğ¸Ñ‚Ğµ Ñ€ĞµĞ¶Ğ¸Ğ¼: ").strip()
        
        if choice == '1.1':
            plot_angle_comparison()
        elif choice == '1.2':
            plot_velocity_comparison()
        elif choice == '2':
            personal_simulation()
        elif choice == '0':
            print("Ğ’Ñ‹Ñ…Ğ¾Ğ´ Ğ¸Ğ· Ğ¿Ñ€Ğ¾Ğ³Ñ€Ğ°Ğ¼Ğ¼Ñ‹...")
            break
        else:
            print("ĞĞµĞ²ĞµÑ€Ğ½Ñ‹Ğ¹ Ğ²Ñ‹Ğ±Ğ¾Ñ€, Ğ¿Ğ¾Ğ¿Ñ€Ğ¾Ğ±ÑƒĞ¹Ñ‚Ğµ ÑĞ½Ğ¾Ğ²Ğ°")

if __name__ == "__main__":
    main()
```
